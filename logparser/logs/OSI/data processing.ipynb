```python
#first_words = df1.apply(lambda x: x.str.split().str[0])

import re

df['Message'] = df['Message'].astype(str)
df['Message'] = df['Message'].apply(lambda x: re.sub('\s+', ' ', x).strip())
df['Message'] = df['Message'].apply(lambda x: x.replace("'", ""))
df['Message'] = df['Message'].apply(lambda x: x.replace(",", ""))
#df['Message'] = df['Message'].apply(lambda x: x[:200])
#df['Message'] = df['Message'].apply(lambda x: x.encode('ascii',errors='ignore'))
```
